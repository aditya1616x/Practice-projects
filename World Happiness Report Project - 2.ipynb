After this all steps now its time for train test split and finding the best model .


```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state= 1)
```

```python
#importing the models to train with
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from scipy.stats import randint
```

```python
models = {
    'DecisionTreeRegressor': DecisionTreeRegressor(random_state=1),
    'LinearRegression': LinearRegression(),
    'SVR': SVR(),
    'RandomForestRegressor': RandomForestRegressor(random_state=1),
    'GradientBoostingRegressor': GradientBoostingRegressor(random_state=1)
}
```

```python
#BULDING A FUNCATION TO SELECT THE BEST MODEL 
def evaluate_models():
    for model_name, model_instance in models.items():
        model_instance.fit(x_train, y_train)  # Fit the model
        y_pred = model_instance.predict(x_test)  # Predict on test data
        mse = mean_squared_error(y_test, y_pred)  # Calculate MSE
        print('Mean Squared Error for', model_name , "=" , mse)
        
```

```python
evaluate_models()
```

# Hyper parameter Tunning

```python
model = DecisionTreeRegressor(random_state=1)
```

```python
from sklearn.model_selection import RandomizedSearchCV , cross_val_score
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'criterion': ['friedman_mse', 'absolute_error', 'poisson', 'squared_error'],
}
```

```python
DTR = RandomizedSearchCV(model , param_grid , cv = 5 , n_jobs = -1)
DTR.fit(x_train,y_train)
```

```python
best_parameters = DTR.best_params_
best_parameters
```

```python
#Training again with the best parameters
Best_parameters = {'criterion': 'absolute_error',
 'max_depth': 20,
 'min_samples_leaf': 3,
 'min_samples_split': 10}
```

```python
Best_parameters = DecisionTreeRegressor(**best_parameters)
```

```python
Best_parameters.fit(x_train,y_train)
y_pred = Best_parameters.predict(x_test)
```

```python
mse = mean_squared_error(y_test , y_pred)
print('Mean Squared Error for DecisionTreeRegressor after hyperparameter tuning:', mse)
```
